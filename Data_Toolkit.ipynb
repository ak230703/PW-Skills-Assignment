# Ques 2. How does broadcasting work in NumPy?
   -  **Broadcasting** in NumPy is the process of automatically expanding arrays of different shapes so that element-wise operations can be performed without explicitly copying data.

It allows you to write concise, vectorized code instead of manual loops.

**- Broadcasting Rules**

When operating on two arrays, NumPy compares their shapes element by element:

1. If the dimensions are equal, they are compatible.

2. If one dimension is 1, it can be stretched (broadcasted) to match the other.

3. If dimensions differ and neither is 1, they are not compatible → error.

- **Broadcasting is Useful**

1. Eliminates explicit loops → cleaner and faster code.

2. Saves memory (no need to copy data multiple times).

3. Essential in machine learning, image processing, and scientific computing

# Ques 4. Explain the use of the groupby() method in Pandas?
   - The **groupby() method** is used to split data into groups based on one or more columns, and then apply some operation (like sum, mean, count, etc.) to each group.

It follows a “split → apply → combine” strategy:

**1. Split →** Divide data into groups based on column(s).

**2. Apply →** Apply a function (aggregation, transformation, or filtering).

**3. Combine →** Merge results back into a new DataFrame.

```python
# Example
import pandas as pd

# Sample data
data = {
    "Department": ["HR", "IT", "HR", "Finance", "IT", "Finance"],
    "Employee": ["Alice", "Bob", "Charlie", "David", "Eve", "Frank"],
    "Salary": [50000, 60000, 52000, 70000, 65000, 72000]
}

df = pd.DataFrame(data)

print("Original DataFrame:")
print(df)

# Group by Department and calculate mean salary
grouped = df.groupby("Department")["Salary"].mean()

print("\nAverage Salary by Department:")
print(grouped)
```

# Ques 6. What are the differences between NumPy arrays and Python lists?  
**1.  Data Structure Type**

 | Feature    | Python List                    | NumPy Array                          |
| ---------- | ------------------------------ | ------------------------------------ |
| Type       | Built-in Python data type      | `numpy.ndarray`                      |
| Dimensions | Can be nested (lists of lists) | Can be N-dimensional (1D, 2D, 3D, …) |

**2. Homogeneity of Elements**

- **Python List →** Can store elements of different types (int, float, string, etc.) in the same list.

- **NumPy Array →** Stores elements of the same data type; converts automatically if types differ.

-  Example:

        import numpy as np

        lst = [1, 2, "3"] # Python list

        arr = np.array([1, 2, 3]) # NumPy array

**3. Performance**

- **Python List →** Slower for mathematical operations, especially large data.

- **NumPy Array →** Faster, optimized for vectorized operations.

- Example
   
          # Element-wise addition
          lst = [1, 2, 3]
          arr = np.array([1, 2, 3])

          # Python list: needs a loop
          lst_sum = [x + 1 for x in lst]

          # NumPy array: vectorized
          arr_sum = arr + 1

**4. Memory Efficiency**

- **Python List →** Stores objects with extra memory overhead.

- **NumPy Array →** Contiguous memory allocation → more memory-efficient.

**5. Operations**

| Operation              | Python List               | NumPy Array                                                |
| ---------------------- | ------------------------- | ---------------------------------------------------------- |
| Addition               | Concatenates lists        | Element-wise addition                                      |
| Multiplication         | Repeats list              | Element-wise multiplication                                |
| Mathematical functions | Must use loops or `map()` | Use NumPy functions (`np.sum`, `np.mean`, `np.sqrt`, etc.) |

- Example:

            lst = [1, 2, 3]
            arr = np.array([1, 2, 3])

            print(lst + lst)  # [1,2,3,1,2,3]
            print(arr + arr)  # [2 4 6]

**6. Slicing and Indexing**

- Both support indexing and slicing.

- NumPy arrays support advanced indexing, boolean indexing, and fancy slicing, which lists don’t.

- Example:

         arr = np.array([10, 20, 30, 40])
         print(arr[arr > 20])  # [30 40]

# Ques 8. What does the term “vectorized operation” mean in NumPy?

In NumPy, a **Vectorized Operation** is a computation that is performed directly on whole arrays (vectors/matrices) at once, without using explicit Python loops.

- Operations are applied element-wise across arrays automatically.

- Under the hood, these operations are implemented in optimized C code, making them much faster than standard Python loops.

**- Why Vectorization is Important**

1. **Performance:** Avoids slow Python for loops.

2. **Conciseness:** One line of code can replace multiple loops.

3. **Readability:** Code is cleaner and easier to understand.

```python
# Example 1: Element-wise Addition

import numpy as np

# Two arrays
a = np.array([1, 2, 3, 4])
b = np.array([10, 20, 30, 40])

# Vectorized addition
c = a + b
print(c)

# Example 2: Multiplying by a Scalar

arr = np.array([1, 2, 3, 4])
result = arr * 10
print(result)

# Example 3: Using a Function
arr = np.array([1, 2, 3, 4])
result = np.sqrt(arr)  # Square root applied element-wise
print(result)

# Comparison with Python Loop - Vectorized operations are faster and cleaner than the loop version, especially for large arrays.
# Using a Python loop
lst = [1, 2, 3, 4]
res = [x + 10 for x in lst]  # slower for large arrays
```

# Ques 13. Why is handling missing data important in Pandas?
   -  **Handling missing data** is a critical step in data analysis and machine learning. Here’s why, specifically in the context of Pandas:

- Why Missing Data Matters

1. **Accurate Analysis**

- Missing values (NaN in Pandas) can skew statistics like mean, median, standard deviation, and correlations.

- **Example:** calculating average salary with missing entries will give a wrong result if not handled.

2. **Prevent Errors**

- Many Pandas operations, as well as ML algorithms, cannot handle NaN values directly and may raise errors.

3. **Consistency in Data**

- Missing data can break visualizations, aggregations, or joins between DataFrames.

- **Example:** plotting with NaN values can cause gaps in graphs.

4. **Better Machine Learning Models**

- ML algorithms usually require complete datasets.

- Handling missing data by imputation or removal improves model performance and prevents biases.

5. **Data Cleaning & Quality**

- Handling missing data is part of data cleaning, which ensures trustworthy insights from the dataset.

```python
# Common Techniques in Pandas
# 1. Detect Missing Data

import pandas as pd
import numpy as np

df = pd.DataFrame({"Age": [25, np.nan, 35, 40], "Salary": [50000, 60000, np.nan, 80000]})
print(df.isnull())

# 2. Drop Missing Values
df_clean = df.dropna()  # Removes rows with any NaN

# 3. Fill Missing Values
df_filled = df.fillna({"Age": df["Age"].mean(), "Salary": df["Salary"].median()})
```

# Ques 15. How does NumPy handle multidimensional arrays?

**1. A multidimensional array (or ndarray) is an array with more than one dimension.**

- Each dimension is called an axis:

a. 1D → vector

b. 2D → matrix (rows × columns)

c. 3D → tensor (e.g., a stack of matrices)

d. N-D → higher-dimensional arrays

**2. Creating Multidimensional Arrays**

**- Example**

                 import numpy as np

                 # 2D array (matrix)
                 arr_2d = np.array([[1, 2, 3],
                                    [4, 5, 6]])

                 # 3D array (tensor)
                 arr_3d = np.array([[[1,2], [3,4]],
                                     [[5,6], [7,8]]])

                print("2D Array:\n", arr_2d)
                print("Shape:", arr_2d.shape)

                print("\n3D Array:\n", arr_3d)
                print("Shape:", arr_3d.shape)


**4. Operations on Multidimensional Arrays**

**- Element-wise operations:** +, -, *, /, etc. applied to all elements.

**- Matrix multiplication:** np.dot() or @ operator for 2D arrays.

**- Axis-specific operations:**

- Sum along rows or columns:

                   np.sum(arr_2d, axis=0)  # sum of columns
                   np.sum(arr_2d, axis=1)  # sum of rows


- Indexing and slicing is supported in multiple dimensions:

                  print(arr_2d[1, 2])  # element at row 1, column 2 → 6
                  print(arr_3d[1, :, 0])  # first column of second matrix → [5,7]

**5. Benefits of NumPy Multidimensional Arrays**

1. Efficient storage → contiguous memory allocation.

2. Vectorized operations → fast computations without loops.

3. Flexible reshaping → reshape(), ravel(), transpose().

4. Broadcasting → allows arithmetic operations on arrays of different shapes.

                    # Reshape 1D array to 2D

# Ques 21. Why is NumPy’s array slicing faster than Python’s list slicing?
   -  NumPy’s array slicing is faster than Python’s list slicing:

**1. Contiguous Memory Storage**

- NumPy arrays are stored in contiguous blocks of memory (like C arrays).

- Python lists are arrays of pointers to objects scattered in memory.

- Contiguous storage allows NumPy to access elements efficiently using simple pointer arithmetic.

**2. Homogeneous Data Type**

- NumPy arrays store elements of the same data type, which allows vectorized operations and direct memory access.

- Python lists can contain mixed data types, so slicing requires type checking and pointer dereferencing for each element.

**3. Slicing Creates Views, Not Copies**

- In NumPy, array slices are views, not copies.

- No new memory is allocated; slicing just adjusts start, stop, and stride pointers.

- Python list slicing creates a new list and copies elements, which is slower.

- Example:

                    import numpy as np

                    arr = np.arange(1000000)
                    lst = list(range(1000000))

                    # NumPy slicing (view, fast)
                    arr_slice = arr[100:200]

                    # Python list slicing (copy, slower)
                    lst_slice = lst[100:200]

**4. Optimized C Implementation**

- NumPy is implemented in C, and slicing operations are highly optimized at a low level.

- Python lists slicing happens in Python interpreter, which is slower.

# Ques 4. How do you calculate the correlation matrix using Seaborn and visualize it with a heatmap?


```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Create a sample DataFrame
data = {
    "Age": [25, 30, 35, 40, 45],
    "Salary": [50000, 60000, 70000, 80000, 90000],
    "Experience": [1, 3, 5, 7, 9]
}

df = pd.DataFrame(data)

# Calculate correlation matrix
corr_matrix = df.corr()

print("Correlation Matrix:\n", corr_matrix)

# Visualize with a heatmap
plt.figure(figsize=(6, 4))
sns.heatmap(corr_matrix, annot=True, cmap="coolwarm", linewidths=0.5)

plt.title("Correlation Matrix Heatmap")
plt.show()
```
